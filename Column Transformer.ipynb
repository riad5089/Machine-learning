# 1.Aam Zindagi

```python
si=SimpleImputer()
x_train_fever=si.fit_transform(x_train[["fever"]])
x_test_fever=si.transform(x_test[["fever"]])
```

```python
x_test_fever
```

```python
oe=OrdinalEncoder(categories=[["Mild","Strong"]])
x_train_cough=oe.fit_transform(x_train[["cough"]])
x_test_cough=oe.transform(x_test[["cough"]])

x_train_cough.shape
```

```python
ohe=OneHotEncoder(drop="first",sparse=False)
x_train_city=ohe.fit_transform(x_train[["gender","city"]])
x_test_city=ohe.transform(x_test[["gender","city"]])
```

```python
x_train_city
```

```python
x_train_transformed=np.concatenate((x_train[["age"]],x_train_fever,x_train_city,x_train_cough),axis=1)
x_test_transformed=np.concatenate((x_test[["age"]],x_test_fever,x_test_city,x_test_cough),axis=1)
```
